#Dataset #1

```python
import geopandas as gpd
```

```python
#loading dataset in
df1 = gpd.read_file('/Hospitalization_Discharge_Rates.geojson')
df1
```

```python
# missing values
missing_values = df1.isnull().sum()
missing_values
```

```python
# column names
df1.columns
```

```python
# map of Hospitalization Discharge Rates for Asthma in Lake County, Illinois
df1.plot("Asthma", legend=True)
```

## Interpretation
### This dataset contains information on Hospitalization Discharge Rates in Lake County, Illinois among different kind of disease. For the purpose of this assignment I chose Asthma and based on the data visualization, the north-east part of Lake County has a greater number of discharge rate for Asthma patients.
